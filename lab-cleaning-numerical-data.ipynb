### 7. Define a function that differentiates between continuous and discrete variables. **Hint**: A number of unique values might be useful. Store continuous data into a `continuous` variable and do the same for `discrete` and categorical.

```python
customer_df.head()
```

```python
data_num = customer_df.select_dtypes(include=[np.number])
data_num.shape
```

```python
data_num.head()
```

```python
data_cat = customer_df.select_dtypes(include=[np.object])
data_cat.shape
```

```python
def cont_disc(arg):
    uniquevalues = len(arg.unique())
    if uniquevalues > 20:
        x = ['continuous',arg.name]
        return x
    else:
        x = ['discrete',arg.name]
        return x
```

```python
data_cat = customer_df
data_num2 = data_num
for i in range(data_num.shape[1]):
    if cont_disc(data_num[data_num.columns[i]])[0] == 'continuous':
        data_cat = data_cat.drop(cont_disc(data_num[data_num.columns[i]])[1], axis=1)
    elif cont_disc(data_num[data_num.columns[i]])[0] == 'discrete':
        data_num2 = data_num2.drop(cont_disc(data_num[data_num.columns[i]])[1], axis=1)
```

### 8. For the categorical data, check if there is some kind of text in a variable so we would need to clean it. **Hint**: Use the same method you used in step 7. Depending on the implementation, decide what to do with the variables you get.

```python
# Shows the unique values for a given column (dataframe series) in case they are less than 20
def column_uniques(arg):
    v = len(arg.unique())
    x = [arg.name, str(v) + ' values']
    if v < 20:
        for i in range(len(arg.unique())):
            x.append(arg.unique()[i])
    return x
```

```python
for i in range(data_cat.shape[1]):
    print(column_uniques(data_cat[data_cat.columns[i]]))
```
